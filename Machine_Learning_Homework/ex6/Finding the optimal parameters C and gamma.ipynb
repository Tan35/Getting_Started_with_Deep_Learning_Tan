## 寻找最优参数 C 和gamma
+ 数据集：data/ex6data3.mat

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn.svm import SVC
```

```python
mat = sio.loadmat('data/ex6data3.mat')
X, y = mat['X'], mat['y'] # 训练集
Xval, yval = mat['Xval'], mat['yval'] # 验证集
```

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1], c=y.flatten(),cmap='jet')
    plt.xlabel('x1')
    plt.ylabel('y1')
```

```python
plot_data()
```

```python
Cvalues = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100] #9
gammas = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100] #9
```

```python
best_score = 0
best_params = (0,0)

for c in Cvalues:
    for gamma in gammas:
        svc = SVC(C=c, kernel='rbf', gamma=gamma)
        svc.fit(X,y.flatten())
        score = svc.score(Xval,yval.flatten())
        if score > best_score:
            best_score = score
            best_params = (c,gamma)
print(best_score, best_params)
```
