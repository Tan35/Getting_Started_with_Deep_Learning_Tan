## 线性可分向量机
+ 任务：观察 C 取值对决策边界的影响
+ 数据集：data/ex6data1.mat

![image.png](<https://i.typlog.com/tanxy/8336257051_772898.png>)

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
data = sio.loadmat('data/ex6data1.mat')
```

```python
data.keys()
```

```python
X, y = data['X'], data['y']
X.shape, y.shape
```

```python
def plot_data():
    plt.scatter(X[:,0],X[:,1],c=y.flatten(),cmap='jet') # 根据 y 的取值来设置点的颜色，1 或者 -1
    plt.xlabel('x1')
    plt.ylabel('y1')
```

```python
plot_data()
```

```python
from sklearn.svm import SVC
```

![image.png](<https://i.typlog.com/tanxy/8336257772_628551.png>)

```python
svc1 = SVC(C=1, kernel='linear')
svc1.fit(X,y.flatten())
```

```python
svc1.predict(X)
```

```python
svc1.score(X,y.flatten())
```

```python
def plot_boundary(model):
    x_min, x_max = -0.5, 4.5
    y_min, y_max = 1.3, 5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    z = model.predict(np.c_[xx.flatten(), yy.flatten()])
    
    zz = z.reshape(xx.shape)
    plt.contour(xx, yy, zz)
```

```python
plot_boundary(svc1)
plot_data()
```

```python
#  改变 C 的取值，看看决策边界的效果
svc100 = SVC(C=100,kernel='linear')
svc100.fit(X,y.flatten())
```

```python
svc100.predict(X)
```

```python
svc100.score(X,y.flatten())
```

```python
plot_boundary(svc100)
plot_data()
```
